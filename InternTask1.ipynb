```python
# ####### Task 1: Exploring and Visualizing the Iris Dataset

# ## Problem Statement
# The goal of this task is to understand how to load, inspect, summarize, and visualize a dataset using Python.

# ## Dataset
# - **Name**: Iris Dataset
# - **Source**: Built into Seaborn 
# - **Description**: Contains measurements of 150 iris flowers from 3 species:
#   - Setosa
#   - Versicolor
#   - Virginica
# - Each sample has 4 features:
#   - Sepal Length
#   - Sepal Width
#   - Petal Length
#   - Petal Width

# ## Tools Used
# - `pandas`: For data loading and inspection
# - `matplotlib` and `seaborn`: For data visualization
```

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#### Dataset Understanding and Description

# We will:
# - Load the dataset
# - Check its structure (number of rows/columns)
# - View the first few entries
# - Understand data types and summary statistics
```

```python
# Load the Iris dataset
df = sns.load_dataset('iris')

# Display basic structure
print("Dataset Shape (rows, columns):", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nData Types:")
print(df.dtypes)
print("\nFirst 5 Rows:")
df.head()
# Get summary statistics for numerical columns
print("Summary Statistics:")
df.describe()
```

```python
#### Data Cleaning and Preparation

# In this case, the Iris dataset is clean and ready to use:
# - No missing values
# - All data is numerical except `species`
# - No duplicates or errors

# So we proceed directly to EDA.
```

```python
#### Exploratory Data Analysis (EDA)

# We create three types of visualizations:
# 1. **Scatter Plot**: To see relationships between variables
# 2. **Histogram**: To understand distribution of a feature
# 3. **Box Plot**: To detect outliers and spread
```

```python
### Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', palette='Set1')
plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
### Histogram Plot
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='sepal_length', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
### Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='sepal_width')
plt.title('Box Plot of Sepal Width')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
